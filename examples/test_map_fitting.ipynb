```python
import pickle
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
```

```python
mapdata = pickle.load(open("example_maps/doh_d2o_gas.pckl",'rb'))
```

```python
w01 = mapdata.w01
Eproj = mapdata.Eproj
dmu = mapdata.dmu/mapdata.dmu[-1]
```

```python
def fit_quad(x, a, b, c):
    return a + b*x + c*x**2
```

```python
popt, pcov = curve_fit(fit_quad, Eproj, w01)
print(popt)
ev_space = np.linspace(0, 0.1, 100)
w01_yv_fit = fit_quad(ev_space, *popt)

fig = plt.figure()
plt.scatter(Eproj, w01)
plt.plot(ev_space, w01_yv_fit, 'r')
plt.xlabel('Eproj (au)')
plt.ylabel('w01 (au)')
plt.show()
```

```python
popt, pcov = curve_fit(fit_quad, Eproj, dmu)
print(popt)
ev_space = np.linspace(0, 0.1, 100)
dmu_yv_fit = fit_quad(ev_space, *popt)

fig = plt.figure()
plt.scatter(Eproj, dmu)
plt.plot(ev_space, dmu_yv_fit, 'r')
plt.xlabel('Eproj (au)')
plt.ylabel('dmu (au)')
plt.show()
```
